```python
import numpy
import random
from matplotlib import pyplot
from matplotlib import colors
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
import datetime
import cartopy.crs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
from PIL import Image
import pandas
```

```python
# Use Pandas to Read Ten Min Sustained Windspeed Data File
Windspeed_File = pandas.read_csv("Ten_Min_Windspeed_Data_Output.csv")
Windspeed_File.head()
```

```python
# Remove Windspeed Stations With Less Than 30 Data Points
Windspeed_Stations_Remove = []
for i in range(len(Windspeed_File.columns)-3):
    Count_Not_Nan = 0
    for j in range(len(Windspeed_File[Windspeed_File.columns[i+2]])):
        if Windspeed_File[Windspeed_File.columns[i+2]][j] >= 0:
            Count_Not_Nan += 1
    print (Windspeed_File.columns[i+2], Count_Not_Nan)
    if Count_Not_Nan < 30:
        Windspeed_Stations_Remove.append(Windspeed_File.columns[i+2])
print (Windspeed_Stations_Remove)
for n in range(len(Windspeed_Stations_Remove)):
    del Windspeed_File[Windspeed_Stations_Remove[n]]
print (Windspeed_File.columns)
print (len(Windspeed_File.columns))
```

```python
# Convert Windspeed Stations Columm Names Into List
Windspeed_Stations_List = list(Windspeed_File.columns)
del Windspeed_Stations_List[0]
del Windspeed_Stations_List[0]
del Windspeed_Stations_List[len(Windspeed_Stations_List)-1]
print (Windspeed_Stations_List)
print ("")

# Input Windspeed Data Into 2D Array
Windspeed_Array = numpy.zeros((len(Windspeed_File),len(Windspeed_File.columns)-3))
for i in range(len(Windspeed_Array)):
    for j in range(len(Windspeed_Array[i])):
        Windspeed_Array[i][j] = Windspeed_File.iloc[i,j+2]
print (Windspeed_Array)
```

```python
# Use Pandas to Read Typhoon CPA Data File
Typhoon_CPA_File = pandas.read_csv("Typhoon_CPA_Data.csv")
Typhoon_CPA_File.head()
```

```python
# Create Lists or Arrays of Each Variable
Typhoon_Code_List = list(Typhoon_CPA_File["Typhoon Code"])
Typhoon_Name_List = list(Typhoon_CPA_File["Typhoon Name"])
Typhoon_Signal_Array = numpy.array(Typhoon_CPA_File["Highest Typhoon Signal Issued"])
CPA_Distance_Array = numpy.array(Typhoon_CPA_File["Closest Point of Approach (km)"])
CPA_Bearing_List = list(Typhoon_CPA_File["CPA Bearing"])
CPA_Intensity_Array = numpy.array(Typhoon_CPA_File["CPA Intensity (km/h)"])
CPA_R_Max_Array = numpy.array(Typhoon_CPA_File["CPA Radius of Maximum Winds (km)"])
print (Typhoon_Code_List)
print (Typhoon_Name_List)
print (Typhoon_Signal_Array)
print (CPA_Distance_Array)
print (CPA_Bearing_List)
print (CPA_Intensity_Array)
print (CPA_R_Max_Array)
```

```python
# Create Empty Lists To Store Typhoon Codes, Names and Windspeeds In Each Quadrant of CPA Directions
Quadrant_Typhoon_Code = [[], [], [], []]
Quadrant_Typhoon_Name = [[], [], [], []]
Quadrant_Windspeeds = [[], [], [], []]
#
# Create Directions Circle List
Directions_Circle = ["N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE", "S", "SSW", "SW", "WSW", "W", "WNW", "NW", "NNW"]
#
# Store Windspeeds According To Each Quadrant of CPA Directions
for i in range(len(CPA_Bearing_List)):
    if CPA_Bearing_List[i] == Directions_Circle[2] or CPA_Bearing_List[i] == Directions_Circle[3] or \
    CPA_Bearing_List[i] == Directions_Circle[4] or CPA_Bearing_List[i] == Directions_Circle[5]:
        Quadrant_Typhoon_Code[0].append(Typhoon_Code_List[i])
        Quadrant_Typhoon_Name[0].append(Typhoon_Name_List[i])
        Quadrant_Windspeeds[0].append(Windspeed_Array[i])
    if CPA_Bearing_List[i] == Directions_Circle[6] or CPA_Bearing_List[i] == Directions_Circle[7] or \
    CPA_Bearing_List[i] == Directions_Circle[8] or CPA_Bearing_List[i] == Directions_Circle[9]:
        Quadrant_Typhoon_Code[1].append(Typhoon_Code_List[i])
        Quadrant_Typhoon_Name[1].append(Typhoon_Name_List[i])
        Quadrant_Windspeeds[1].append(Windspeed_Array[i])
    if CPA_Bearing_List[i] == Directions_Circle[10] or CPA_Bearing_List[i] == Directions_Circle[11] or \
    CPA_Bearing_List[i] == Directions_Circle[12] or CPA_Bearing_List[i] == Directions_Circle[13]:
        Quadrant_Typhoon_Code[2].append(Typhoon_Code_List[i])
        Quadrant_Typhoon_Name[2].append(Typhoon_Name_List[i])
        Quadrant_Windspeeds[2].append(Windspeed_Array[i])
    if CPA_Bearing_List[i] == Directions_Circle[14] or CPA_Bearing_List[i] == Directions_Circle[15] or \
    CPA_Bearing_List[i] == Directions_Circle[0] or CPA_Bearing_List[i] == Directions_Circle[1]:
        Quadrant_Typhoon_Code[3].append(Typhoon_Code_List[i])
        Quadrant_Typhoon_Name[3].append(Typhoon_Name_List[i])
        Quadrant_Windspeeds[3].append(Windspeed_Array[i])
print (Quadrant_Typhoon_Code)
print (" ")
print (Quadrant_Typhoon_Name)
print (" ")
print (Quadrant_Windspeeds)
```

```python
# Create Empty Arrays to Store Windspeed Data of Each Quadrant With Windspeed Station at Independent Variable
East_Quadrant_Windspeeds = numpy.zeros((len(Windspeed_Stations_List),len(Quadrant_Windspeeds[0])))
South_Quadrant_Windspeeds = numpy.zeros((len(Windspeed_Stations_List),len(Quadrant_Windspeeds[1])))
West_Quadrant_Windspeeds = numpy.zeros((len(Windspeed_Stations_List),len(Quadrant_Windspeeds[2])))
North_Quadrant_Windspeeds = numpy.zeros((len(Windspeed_Stations_List),len(Quadrant_Windspeeds[3])))
#
# Input Data From Previous 3D List Into Arrays
for k in range(4):
    for i in range(len(Quadrant_Windspeeds[k])):
        for j in range(len(Windspeed_Stations_List)):
            if k == 0:
                East_Quadrant_Windspeeds[j][i] = Quadrant_Windspeeds[k][i][j]
            if k == 1:
                South_Quadrant_Windspeeds[j][i] = Quadrant_Windspeeds[k][i][j]
            if k == 2:
                West_Quadrant_Windspeeds[j][i] = Quadrant_Windspeeds[k][i][j]
            if k == 3:
                North_Quadrant_Windspeeds[j][i] = Quadrant_Windspeeds[k][i][j]
print (East_Quadrant_Windspeeds)
print (" ")
print (South_Quadrant_Windspeeds)
print (" ")
print (West_Quadrant_Windspeeds)
print (" ")
print (North_Quadrant_Windspeeds)
```

```python
# Calculate The Ratio Between Windspeed at Cheung Chau and Windspeed at Each Station for Each Quadrant
East_Quadrant_Windspeed_Ratio_To_CC = East_Quadrant_Windspeeds / East_Quadrant_Windspeeds[2]
print (East_Quadrant_Windspeed_Ratio_To_CC)
print ("")
South_Quadrant_Windspeed_Ratio_To_CC = South_Quadrant_Windspeeds / South_Quadrant_Windspeeds[2]
print (South_Quadrant_Windspeed_Ratio_To_CC)
print ("")
West_Quadrant_Windspeed_Ratio_To_CC = West_Quadrant_Windspeeds / West_Quadrant_Windspeeds[2]
print (West_Quadrant_Windspeed_Ratio_To_CC)
print ("")
North_Quadrant_Windspeed_Ratio_To_CC = North_Quadrant_Windspeeds / North_Quadrant_Windspeeds[2]
print (North_Quadrant_Windspeed_Ratio_To_CC)
print ("")
```

```python
# Calculate the Average Ratio Between Windspeed at Cheung Chau and Windspeed at Each Station for Each Quadrant
Windspeed_Ratio_Stations_To_CC = numpy.zeros((4, len(Windspeed_Stations_List)))
for k in range(4):
    for j in range(len(Windspeed_Stations_List)):
        if k == 0:
            Windspeed_Ratio_Stations_To_CC[k][j] = numpy.nanmean(East_Quadrant_Windspeed_Ratio_To_CC[j])
        if k == 1:
            Windspeed_Ratio_Stations_To_CC[k][j] = numpy.nanmean(South_Quadrant_Windspeed_Ratio_To_CC[j])
        if k == 2:
            Windspeed_Ratio_Stations_To_CC[k][j] = numpy.nanmean(West_Quadrant_Windspeed_Ratio_To_CC[j])
        if k == 3:
            Windspeed_Ratio_Stations_To_CC[k][j] = numpy.nanmean(North_Quadrant_Windspeed_Ratio_To_CC[j])
print (Windspeed_Ratio_Stations_To_CC)
```

```python
# Find Average of the Average Ratios Between Windspeed at Cheung Chau and Windspeed at Each Station for Each Quadrant
Windspeed_Ratio_Average_To_CC = numpy.zeros(4)
for k in range(4):
    Windspeed_Ratio_Average_To_CC[k] = numpy.mean(Windspeed_Ratio_Stations_To_CC[k])
print (Windspeed_Ratio_Average_To_CC)
```

```python
# Find Average Ratios Between Windspeed at Each Station and Windspeed at Average of HK
Windspeed_Ratio_Stations_To_HK_Ave = numpy.zeros((4, len(Windspeed_Stations_List)))
for k in range(4):
    for j in range(len(Windspeed_Stations_List)):
        Windspeed_Ratio_Stations_To_HK_Ave[k][j] = Windspeed_Ratio_Stations_To_CC[k][j] / \
        Windspeed_Ratio_Average_To_CC[k]
print (Windspeed_Ratio_Stations_To_HK_Ave)
```

```python
# Create Output File For Windspeed Data
Quadrants_List = ["Eastern", "Southern", "Western", "Northern"]
Windspeed_Ratio_Output_File = open('Windspeed_Ratio_To_HK_Average_Output.csv','w')
for k in range(5):
# Insert Heading With Quadrant and Windspeed Stations Name
    if k == 0:
        for j in range(len(Windspeed_Stations_List)+1):
            if j == 0:
                Windspeed_Ratio_Output_File.write("%s" % "Quadrant")
                Windspeed_Ratio_Output_File.write("%s" % ",")
            else:
                Windspeed_Ratio_Output_File.write("%s" % Windspeed_Stations_List[j-1])
                Windspeed_Ratio_Output_File.write("%s" % ",")
#Insert Windspeed Ratio Data For Each Station and Each Quadrant
    else:
        for j in range(len(Windspeed_Stations_List)+1):
            if j == 0:
                Windspeed_Ratio_Output_File.write("%s" % Quadrants_List[k-1])
                Windspeed_Ratio_Output_File.write("%s" % ",")
            else:
                Windspeed_Ratio_Output_File.write("%s" % Windspeed_Ratio_Stations_To_HK_Ave[k-1][j-1])
                Windspeed_Ratio_Output_File.write("%s" % ",")
    Windspeed_Ratio_Output_File.write("%s\n" % " ")
#
# Close File
Windspeed_Ratio_Output_File.close()
print ("Output is saved into file. File closed.")
print
```
